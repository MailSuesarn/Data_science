## Feature engineering


Well perform this feature engineering step first, as the original feature will be removed afterward to avoid redundancy. This idea is computing the average SalePrice across groups of interest, known as **target encoding**. We calculate these averages using the training set and apply them to the test set. For categories missing in the test set or unknown future values, we impute using the overall mean from the training data to ensure robustness.

We should avoid computing the average across the entire dataset (even just the training set) before splitting into training and validation sets, as this can lead to **data leakage** and produce misleading validation results.

```python
class TargetMeanEncoder:
    def __init__(self, column, target):
        self.column = column
        self.target = target
        self.mapping = {}
        self.global_mean = None

    def fit(self, X, y):
        df = X.copy()
        df[self.target] = y
        self.global_mean = y.mean()
        self.mapping = df.groupby(self.column)[self.target].mean().to_dict()
        return self

    def transform(self, X):
        encoded = X[self.column].map(self.mapping)
        return encoded.fillna(self.global_mean)

    def fit_transform(self, X, y):
        self.fit(X, y)
        return self.transform(X)
```

To select good features for Target Mean Encoding (TME):

 Choose features with high variance in mean SalePrice across categories,  this shows the feature is predictive.

 Avoid features with low variance, where most categories have similar target means.

 Exclude rare or unbalanced features, where one category dominates or some appear very few times (e.g., < 5), as they may overfit.

In short: Look for high target variance + balanced category distribution.

```python
def analyze_target_encoding_candidate(df, feature, target='SalePrice', top_n=None):
    """
    Analyze a categorical feature to determine if it's a good candidate for Target Mean Encoding.

    Shows:
    - Mean target value per category
    - Count and percentage of each category
    - Optionally limits to top N most frequent categories
    """
    total_count = len(df)

    summary = (
        df.groupby(feature)[target]
        .agg(['mean', 'count'])
        .rename(columns={'mean': f'{target}_mean', 'count': 'count'})
        .sort_values(by=f'{target}_mean', ascending=False)
    )

    summary['%_of_total'] = (summary['count'] / total_count * 100).round(2)

    if top_n:
        summary = summary.head(top_n)

    print(f"Analysis for feature: {feature}")
    return summary
```

```python
class FeatureEngineer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        df = X.copy()

        # House-related ages
        df["HouseAge"] = df["YrSold"] - df["YearBuilt"]
        df["RemodAge"] = df["YrSold"] - df["YearRemodAdd"]
        df["GarageAge"] = df["YrSold"] - df["GarageYrBlt"]

        # Since months are cyclical (December wraps to January), Captures seasonality
        df['MoSold_sin'] = np.sin(2 * np.pi * df['MoSold'] / 12)
        df['MoSold_cos'] = np.cos(2 * np.pi * df['MoSold'] / 12)
        df.drop(columns=["MoSold"], inplace=True)

        # Basement finished area and remove TotalBsmtSF
        df["BsmtFinishedSF"] = df["BsmtFinSF1"] + df["BsmtFinSF2"]
        if "TotalBsmtSF" in df.columns:
            df.drop(columns=["TotalBsmtSF"], inplace=True)

        # House age bins
        df['HouseAgeBin'] = pd.cut(
            df['HouseAge'],
            bins=[0, 10, 30, 80, 1000],
            labels=['New', 'Recent', 'Old', 'Very Old']
        )

        # Quality  Condition score
        df["OverallScore"] = df["OverallQual"] * df["OverallCond"]

        # Remodel flag
        df['HasRemodel'] = (df['YearRemodAdd'] != df['YearBuilt']).astype(int)

        # Total rooms
        room_features = ['TotRmsAbvGrd', 'FullBath', 'HalfBath']
        valid_features = [col for col in room_features if col in df.columns]
        df["TotalRooms"] = df[valid_features].sum(axis=1)

        # Has Misc Feature flag
        df['HasMiscFeature'] = (df['MiscFeature'].fillna('None') != 'None').astype(int)
        df.drop(columns='MiscFeature', inplace=True)

        return df
```

```python
X_train = df_train.drop(columns=['SalePrice'])
X_val = df_val.drop(columns=['SalePrice'])
X_test = df_test.copy()

y_train = df_train['SalePrice']
y_val = df_val['SalePrice']
```

```python
nominal_imputer = SimpleImputer(strategy='constant', fill_value='None')
ordinal_imputer = SimpleImputer(strategy='constant', fill_value='None')

nominal_pipe = Pipeline(steps=[
    ('imputer', nominal_imputer),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

ordinal_pipe = Pipeline(steps=[
    ('imputer', ordinal_imputer),
    ('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])


preprocessor = ColumnTransformer(
    transformers=[
        ('nominal', nominal_pipe, nominal_cols),
        ('ordinal', ordinal_pipe, ordinal_cols)
    ],
    remainder='passthrough'
)


full_pipeline = Pipeline(steps=[
    ('feature_engineering', FeatureEngineer()),
    ('preprocessing', preprocessor)
])
```

```python
full_pipeline.fit(X_train)

X_train_encoded  = full_pipeline.transform(X_train)
X_val_encoded  = full_pipeline.transform(X_val)
X_test_encoded  = full_pipeline.transform(X_test)
```

```python
def check_column_alignment(train_df, val_df, test_df):
    train_cols = set(train_df.columns)
    val_cols = set(val_df.columns)
    test_cols = set(test_df.columns)

    print("=== Column Alignment Check ===\n")

    # Train vs Val
    print("Train vs Val:")
    print("Columns in train but not in val:", sorted(train_cols - val_cols))
    print("Columns in val but not in train:", sorted(val_cols - train_cols))
    print()

    # Train vs Test
    print("Train vs Test:")
    print("Columns in train but not in test:", sorted(train_cols - test_cols))
    print("Columns in test but not in train:", sorted(test_cols - train_cols))
    print()

    # Val vs Test
    print("Val vs Test:")
    print("Columns in val but not in test:", sorted(val_cols - test_cols))
    print("Columns in test but not in val:", sorted(test_cols - val_cols))
    print()

    # Common Columns
    common_cols = train_cols & val_cols & test_cols
    print(f"Columns shared across all sets: {len(common_cols)}")

    return common_cols
```

## Hypothesis testing

This step involves hypothesis testing to determine whether observed patterns are statistically significant, providing a foundation for insights or subsequent analytical decisions.

In this case:

***We aim to test the hypothesis that remodeled houses have a significantly different average sale price compared to unremodeled houses.***

The step is to conduct a two-sample t-test. Recall the steps for conducting a hypothesis test:


1.   State the null hypothesis and the alternative hypothesis
2.   Choose a signficance level
3.   Find the p-value
4.   Reject or fail to reject the null hypothesis



$H_0$: There is no statistically significant difference in the average sale price between remodeled and unremodeled houses.

$H_A$: There is a statistically significant difference in the average sale price between remodeled and unremodeled houses.

Choose 5% as the significance level and proceed with a two-sample t-test.

```python
df_train_vis.groupby('remainder__HasRemodel')['SalePrice'].mean()
```

```python
#hypothesis test, A/B test
#significance level

remodeled = df_train_vis[df_train_vis['remainder__HasRemodel'] == 1]['SalePrice']
unremodeled = df_train_vis[df_train_vis['remainder__HasRemodel'] == 0]['SalePrice']

stats.ttest_ind(a=remodeled, b=unremodeled, equal_var=False)
```

**Note:** Since the p-value is significantly larger than the significance level of 5%, **Fail to reject the null hypothesis**.

There is **no statistically significant difference** in the average sale price between remodeled and unremodeled houses at the 5% significance level.

### Evaluate model

Evaluate the model performance by calculating the residual sum of squares and the explained variance score (R^2). Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

```python
def evaluate_regression_model(model, X_val, y_val, model_name="Model"):

    # Predict using the model
    y_pred = model.predict(X_val)

    # Compute metrics
    r2 = r2_score(y_val, y_pred) # or Coefficient of determination
    mae = mean_absolute_error(y_val, y_pred)
    mse = mean_squared_error(y_val, y_pred)
    rmse = np.sqrt(mse)

    # Display results
    print(f"\nEvaluation results for {model_name}:")
    print(f"R:   {r2:.4f}")
    print(f"MAE:  {mae:.2f}")
    print(f"MSE:  {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")

    return {
        'model': model_name,
        'r2': r2,
        'mae': mae,
        'mse': mse,
        'rmse': rmse
    }
```

```python
def train_tree_model(model_class, param_grid, X_train, y_train, X_val, y_val, random_state=42, verbose=True):
    """
    Perform grid search for a tree-based model using RMSE and R on validation set.

    Args:
        model_class: model constructor (e.g., RandomForestRegressor, XGBRegressor)
        param_grid: dict, hyperparameters to search (as in sklearn's GridSearch)
        X_train, y_train: training data
        X_val, y_val: validation data
        random_state: int, reproducibility
        verbose: bool, whether to print all iterations

    Returns:
        best_model: trained model with best params
        best_params: dict of best hyperparameters
        best_score: best RMSE (lower is better)
    """
    best_score = float('inf')
    best_params = None
    best_model = None

    for params in ParameterGrid(param_grid):
        # Include random_state if supported
        if 'random_state' in model_class().get_params():
            params['random_state'] = random_state

        model = model_class(**params)
        model.fit(X_train, y_train)

        y_pred = model.predict(X_val)

        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        r2 = r2_score(y_val, y_pred)

        if verbose:
            print(f"Params: {params} | RMSE: {rmse:.2f} | R: {r2:.4f}")

        if rmse < best_score:
            best_score = rmse
            best_params = params
            best_model = model

    if verbose:
        print("\nBest Parameters:", best_params)
        print(f"Best Validation RMSE: {best_score:.2f}")

    return best_model, best_params, best_score
```
